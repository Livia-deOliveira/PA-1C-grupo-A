```python
# Escrever um programa que imprime de 5 até 0

for i in range (6):
  print (5 - i)
```

```python
# Escrever um programa que imprime os números pares de 0 a 10

for i in range (11):
  if i % 2 == 0:
    print (i)
```

```python
# Escrever um programa que imrprime ops números impares existentes de 0 a 10

for i in range (11):
  if i % 2 != 0:
    print (i)
```

```python
# Escrever um programa onde o usuário digita um número de 1 a 10 e o programa imprime a tabuada de tal número

numero = int(input("Digite um número de 1 a 10: "))
for  i in range (1, 11):
  print (f" {numero} X {i} = {numero * i}")
```
