```python
# Escrever um programa que faz a 04 operações aritimética. (Calculadora simples)

print ("|---------------CALCUDORA SIMPLES---------------|")
print ("|                                               |")
print ("|----------------------MENU---------------------|")
print ("|                                               |")
print ("| Soma = 1                                      |")
print ("|                                               |")
print ("| Subtração = 2                                 |")
print ("|                                               |")
print ("| Multiplicação = 3                             |")
print ("|                                               |")
print ("| Divisão = 4                                   |")
print ("|-----------------------------------------------|")


# Entrada

numero1 = float (input("Digite o primeiro número: "))
numero2 = float (input("Digite o segundo número: "))
operacao = int (input("Digite o número da operação: "))

# Processamento e Saída

soma = numero1 + numero2
subtracao = numero1 - numero2
multiplicacao = numero1 * numero2
divisao = numero1 / numero2

if operacao == 1:
    print (f"O resultado da soma é: {soma}")

elif operacao == 2:
    print (f"O resultado da subtração é: {subtracao}")

elif operacao == 3:
    print (f"O resultado da multiplicação é: {multiplicacao}")

elif operacao == 4:
    print (f"O resultado da divisão é: {divisao}")

else:
    print ("Operação inválida")
```

```python
# Escrever um programa onde o usuário digita sua idade e o programa retorna se ele é adulto, adolescente ou criança.

# Entrada

nome = str (input("Digite seu nome: "))
idade = int (input("Digite sua idade: "))

# Processamento e Saída

if idade >= 18:
    print (f"{nome}, você é adulto.")

elif idade >= 12:
    print (f"{nome}, você é adolescente.")

else:
        print (f"{nome}, você é criança")
```
