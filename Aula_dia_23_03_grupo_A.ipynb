```python
# Escrever um programa que verifica se um número é par ou ímpar

numero = int (input("Digite um número: "))

if numero % 2 == 0:

  print (f"{numero} é par.")

else:

  print (f"{numero} é ímpar.")
```

```python
# Escrever um programa que verifica uma senha simples, senha 1234, saída acesso antecipado

senha = int (input("Digite sua senha: "))

if senha == 1234:

  print ("Acesso permitido, seja bem-vindo!")

else:

  print ("Senha incorreta, tente novamente.")

  senha = int (input("Digite sua senha: "))

  if senha == 1234:

    print ("Acesso permitido, seja bem-vindo!")

  else:

    print ("Acesso negado, tente novamente mais tarde.")
```

```python
# Escrever um programa que:
# Peça o valor da compra
# Se o valor for maior que 100, aplique um desconto de 10%
# Se não, mostre o valor normal

compra = float (input("Digite o valor da compra: "))

if compra > 100:

  desconto = compra * 0.9
  print (f"O valor da compra com desconto é de R${desconto:.2f}")

else:

  print (f"O valor da compra é de R${compra:.2f}")
```

```python
# Escrever um programa que o aluno digita 04 notas dos bimestres e o programa calcula a média e mostra se o aluno foi aprovado, reprovado ou se ficou de recuperação

nota1 = float (input("Digite a nota do primeiro bimestre: "))
nota2 = float (input("Digite a nota do segundo bimestre: "))
nota3 = float (input("Digite a nota do terceiro bimestre: "))
nota4 = float (input("Digite a nota do quarto bimestre: "))

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 7:

  print (f" ")
  print (f"---------------------------------------------")
  print (f" Você foi aprovado com média {media:.2f}, parabéns!")
  print (f"---------------------------------------------")

elif media >= 5:

  print (f" ")
  print (f"---------------------------------------------")
  print (f"Você ficou de recuperação com média {media:.2f}, essa foi por pouco!")
  print (f"---------------------------------------------")

else:

    print (f" ")
    print (f"---------------------------------------------")
    print (f"Você foi reprovado com média {media:.2f}, se dedique mais na próxima!")
    print (f"---------------------------------------------")
```
