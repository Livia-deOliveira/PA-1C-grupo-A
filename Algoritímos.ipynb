Escrever um Algoritímo e um programa que recebe 04 notas do aluno e calcula a média, exibindo-a após o calculo

# Entrada

Ler as notas de matemática, português, história e inglês

# Processamento

Somar todas as notas e dividir o resultado pela quantidade de elementos (4)

# Saída

Mostrar o resultado da conta

```python
# Entrada

matemática = 8
português = 9
história = 10
inglês = 10

# Processamento

média = (matemática + português + história + inglês) / 4

# Saída

print (f" A nota final do aluno é: {média}")
```
