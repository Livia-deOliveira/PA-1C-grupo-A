```python
# Escrever um programa onde o usuário digita um número e o programa imprime de 0 até o número digitado

num = int(input("Digite um número: "))
for i in range (0, num+1):
  print (i)
```

```python
# Escrever um programa onde o usuário digita um número e o programa imprime em ordem decrescente até o zero

num = int(input("Digite um número: "))
for i in range (num, -1, -1):
  print (i)
```

```python
# Escrever um programa onde o usuário digita um número e o programa imprime os números pares até o número digitado

num = int(input("Digite um número: "))
for i in range (0, num+1, 2):
  print (i)
```

```python
# Escrever um programa onde o usuário digita um número e o programa calcula a soma de 1 até o número digitado

num = int(input("Digite um número: "))
soma = 0
for i in range (1, num+1):
  soma = soma + i
print (soma)
```

```python
# Escrever um programa onde o usuário insere uma palavra e o programa conta os caracteres de tal

palavra = input("Digite uma palavra: ")
print (len(palavra))
```

```python
# Escrever um programa onde o usuário insere uma palavra e o programa conta as vogais da palavra inserida

palavra = input("Digite uma palavra: ")
vogais = "aeiouAEIOU"
contador = 0
for letra in palavra:
  if letra in vogais:
    contador = contador + 1
    print (contador)
```

```python
# Escrever um programa onde o usuário digita 02 números e o programa informa qual é o maior entre eles

num = int(input("Digite um número: "))
num2 = int(input("Digite outro número: "))
if num > num2:
    print (num)
else :
    print (num2)
```
