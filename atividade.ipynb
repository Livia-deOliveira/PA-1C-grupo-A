```python
# Questão 4 - programa de  sistema de energia

vida = int (input("Digite a quantidade de energia do personagem: "))

if vida == 0:

    print (f"O personagem morreu D:")


elif vida >= 1 and vida <= 30:

    print (f"Cuidado! O personagem está em estado crítico.")


elif vida >= 31 and vida <= 70:

    print (f"Vida em estado normal.")


elif vida >= 71 and vida <= 100:
    
    print (f"O personagem está muito bem :)")
```

```python
# Questão 5 - programa de média escolar

nota1 = float (input("insira a primeira nota: "))
nota2 = float (input("insira a segunda nota: "))
nota3 = float (input("insira a terceira nota: "))
nota4 = float (input("insira a quarta nota: "))

media = (nota1 + nota2 + nota3 + nota4) / 4


if media < 5:

    print (f"Sua nota final é {media}. Você foi reprovado.")


elif media >= 5 and media <= 6.9:

    print (f"Sua nota final é {media}. Você está de recuperação.")


elif media >= 7:
    
    print (f"Sua nota final é {media}. Você está aprovado, parabéns!")
```
